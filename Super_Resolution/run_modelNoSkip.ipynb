```python
from Unet_Architecture.Super_Resolution.Library import *
from Unet_Architecture.Super_Resolution.Prepare_Dataset import data
from Unet_Architecture.Super_Resolution.Train_Val import training, valid_epoch, plot_result, predict_and_display
from Unet_Architecture.Super_Resolution.UnetNoSkipConnection import UnetNoSkip
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
device
```

```python
path_train = "/content/Unet_Architecture/Super_Resolution/dataset/dataset/train"
path_val = "/content/Unet_Architecture/Super_Resolution/dataset/dataset/val"
train_dataset, val_dataset, train_loader, val_loader = data(path_train, path_val, IMG_WIDTH, IMG_HEIGHT, BATCH_SIZE)
```

```python
in_channels = 3
num_classes = 3
num_epochs = 100
model = UnetNoSkip(in_channels, num_classes, IMG_HEIGHT, IMG_WIDTH)
model.to(device)

criterion = nn.L1Loss()
optimizer = optim.Adam(model.parameters(), lr = 0.0005, betas=[0.5, 0.999])
model, metrics = training(model, optimizer, criterion, train_loader, val_loader, num_epochs, device)
```

```python
plot_result(num_epochs, metrics['train_psnr'], metrics['val_psnr'],
            metrics['train_losses'], metrics['val_losses'])
```

```python
predict_and_display(model, val_loader, device)
```

```python
test_psnr, test_loss = valid_epoch(model, criterion, val_loader, device)
test_psnr, test_loss
```
